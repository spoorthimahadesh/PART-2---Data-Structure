```python
#Q1
cart = []

# function to add item
def add_to_cart(item_name, quantity):

    # checking if item is there
    if item_name not in menu:
        print(f"{item_name} not in menu")
        return

    # checking availability
    if menu[item_name]["available"] == False:
        print(f"{item_name} is not available now")
        return

    # checking if already in cart
    found = False

    for x in cart:
        if x["item"] == item_name:
            x["quantity"] = x["quantity"] + quantity
            print(f"{item_name} quantity updated to {x['quantity']}")
            found = True
            break

    # if not found then add new
    if found == False:
        new_item = {}
        new_item["item"] = item_name
        new_item["quantity"] = quantity
        new_item["price"] = menu[item_name]["price"]

        cart.append(new_item)

        print(f"{item_name} added")

# testing
add_to_cart("Paneer Tikka", 2)
add_to_cart("Paneer Tikka", 1)
add_to_cart("Chicken Wings", 1)
add_to_cart("Pizza", 1)

print(cart)
```

```python
#Q2
def remove_from_cart(item_name):

    found = False   # assuming not found

    # checking each item in cart
    for x in cart:
        if x["item"] == item_name:
            cart.remove(x)
            print(f"{item_name} removed from cart")
            found = True
            break

    # if not found
    if found == False:
        print(f"{item_name} not in cart")


# testing
remove_from_cart("Paneer Tikka")
remove_from_cart("Dal Tadka")
```

```python
#Q3
def update_quantity(item_name, new_quantity):

    found = False   # assuming item not there

    # checking in cart
    for x in cart:
        if x["item"] == item_name:
            x["quantity"] = new_quantity
            print(f"{item_name} quantity changed to {new_quantity}")
            found = True
            break

    # if not found
    if found == False:
        print(f"{item_name} not in cart")


# testing
add_to_cart("Dal Tadka", 1)
update_quantity("Dal Tadka", 4)
update_quantity("Rasgulla", 2)
```

```python
#Q2
best_day = ""
max_total = 0

# checking each day
for d, orders in sales_log.items():

    total = 0

    for o in orders:
        total = total + o["total"]

    if total > max_total:
        max_total = total
        best_day = d

print(f"Best day is {best_day} with Rs {max_total}")
```

```python
#Q3
# counting items

item_count = {}

for d, orders in sales_log.items():
    for o in orders:
        for i in o["items"]:

            if i in item_count:
                item_count[i] = item_count[i] + 1
            else:
                item_count[i] = 1


# finding most ordered item
max_count = 0
max_item = ""

for i, c in item_count.items():
    if c > max_count:
        max_count = c
        max_item = i

print(f"Most ordered item is {max_item} with {max_count} orders")

print("")
print("All items count :")

for i, c in item_count.items():
    print(f"{i} - {c}")
```
